# Лабораторна робота №4
## Варіант №3
## Бонадревська Поліна, ІПС-41

\begin{cases}
8.2 x_1 + 7.9 x_2 = 123 \\
11.2 x_1 + 16.3 x_2 = 201 \\
-16.3 x_1 - 8.3 x_2 = 169
\end{cases}

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-darkgrid')
```

```python
def equation1(x1):
    return (123 - 8.2*x1) / 7.9

def equation2(x1):
    return (201 - 11.2*x1) / 16.3

def equation3(x1):
    return (-169 + 16.3*x1) / -8.3

x1_values = np.linspace(0, 15, 100)
```

```python
plt.figure(figsize=(15,5), dpi=100)

plt.subplot(1,3,1)
plt.plot(x1_values, equation1(x1_values),  color='IndianRed',  linestyle='--')
plt.title('Equation 1: 8.2x1 + 7.9x2 = 123')
plt.subplot(1,3,2)
plt.plot(x1_values, equation2(x1_values) ,color='Teal', linestyle=':')
plt.title('Equation 2: 11.2x1 +  16.3x2 = 201')
plt.subplot(1,3,3)
plt.plot(x1_values,  equation3(x1_values), color='Fuchsia',  linestyle='-.')
plt.title('Equation 3:  -16.3x1 - 8.3x2 = 169')
plt.tight_layout()
plt.show()
```

```python
x1_values = np.linspace(4, 10, 100)
AB_pairs = [
    (np.array([[8.2, 7.9], [11.2, 16.3]]), np.array([123, 201])),
    (np.array([[8.2, 7.9], [-16.3, -8.3]]), np.array([123, -169])),
    (np.array([[11.2, 16.3], [-16.3, -8.3]]), np.array([201, -169]))
]
solutions = [np.linalg.solve(A, B) for A, B in AB_pairs]
plt.figure(figsize=(10,5), dpi=200)
plt.plot(x1_values, equation1(x1_values), label='8.2x1 + 7.9x2 = 123', color='IndianRed',  linestyle='--')
plt.text(6, equation1(6)-0.5, r'$8.2x_1 + 7.9x_2 = 123$', rotation=-14, color='IndianRed')
plt.plot(x1_values, equation2(x1_values), label='11.2x1 + 16.3x2 = 201', color='Teal', linestyle=':')
plt.text(4, equation2(4)-0.5, r'$11.2x_1 + 16.3x_2 = 201$', rotation=-10, color='Teal')
plt.plot(x1_values,  equation3(x1_values), label='-16.3x1 - 8.3x2 = 169', color='Fuchsia',  linestyle='-.')
plt.text(8, equation3(8)-1.7,  r'$-16.3x_1 - 8.3x_2 = 169$', rotation=-26, color='Fuchsia')
# Вивід точок перетину
for s in solutions:
    plt.scatter(s[0], s[1], color='Navy', marker='o', zorder=5)
    plt.annotate(f'({s[0]:.2f}, {s[1]:.2f})', (s[0], s[1]), textcoords="offset points", xytext=(-10,5), ha='center', fontsize=10, color='Navy')


plt.fill_between(x1_values, np.maximum(np.maximum(equation1(x1_values), equation3(x1_values)), equation3(x1_values)), np.maximum(np.minimum(equation2(x1_values), equation1(x1_values)), equation3(x1_values)), color='gray', alpha=0.5)
plt.xlabel(r'$x_1$', fontsize=12)
plt.ylabel(r'$x_2$', fontsize=12)
plt.title('Intersection of three lines', fontsize=15)
plt.grid(color = 'purple', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.savefig('graph.jpg')
plt.savefig('graph.png')
plt.savefig('graph.svg')
plt.show()
```
